### Step 6.5: setup.py 및 requirements.txt 복사

````python
# setup.py와 requirements.txt 파일 존재 확인
import os

setup_notebook_dir = os.path.expanduser('~/ssl')  # setup.ipynb가 있는 디렉토리
setup_py_path = os.path.join(setup_notebook_dir, 'setup.py')
req_path = os.path.join(setup_notebook_dir, 'requirements.txt')

print("파일 확인 중...")
if os.path.exists(setup_py_path):
    print(f"✅ setup.py 발견: {setup_py_path}")
else:
    print(f"❌ setup.py 없음: {setup_py_path}")

if os.path.exists(req_path):
    print(f"✅ requirements.txt 발견: {req_path}")
else:
    print(f"❌ requirements.txt 없음: {req_path}")
````

````python
# setup.py와 requirements.txt를 CATANet 프로젝트 디렉토리로 복사
import shutil

files_to_copy = [
    ('requirements.txt', req_path),
    ('setup.py', setup_py_path)
]

for dst_file, src_file in files_to_copy:
    if os.path.exists(dst_file):
        print(f"⚠️  {dst_file} 이미 존재합니다. 덮어쓰기 중...")
    
    if os.path.exists(src_file):
        shutil.copy(src_file, dst_file)
        print(f"✅ {dst_file} 복사 완료")
    else:
        print(f"❌ {src_file} 파일을 찾을 수 없습니다.")
````

### Step 6.6: BasicSR 프로젝트 설치 (개발 모드)

````python
# BasicSR을 개발 모드로 설치 (이미 설치되어 있으면 스킵)
import subprocess

# 이미 설치되어 있는지 확인
result = subprocess.run(
    ['~/ssl/ssl_env/bin/pip', 'show', 'basicsr'],
    shell=True,
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print("⚠️  basicsr이 이미 설치되어 있습니다. 재설치 중...")
    !~/ssl/ssl_env/bin/pip install -e . --force-reinstall --no-deps
    print("✅ basicsr 재설치 완료")
else:
    print("basicsr 설치 중...")
    !~/ssl/ssl_env/bin/pip install -e .
    print("✅ basicsr 설치 완료")
````
